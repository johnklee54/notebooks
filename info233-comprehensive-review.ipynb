### Chapter 10 OOP- Object Oriented Programming
#### Object Oriented Programming - OOP

Object Oriented programming is a programming technique which allows developers to build reusable code.  A few new terms to get us started.

We can use OOP to create representations of real objects with code.

What is an object?

**Objects** - we are surrounded by objects in real life. Everything we see is pretty much an object. For example, a person is an object and we can describe a person with attributes which all people have. A person's name and race, A person's height and weight.  

**Class** - In programming we can create and object by using a class. A class is a blueprint for creating objects. Picture a cookie cutter that can cut out a certain type of cookie, A cookie's size, shape etc... We can use the cookie cutter to continue to stamp out as many cookies as we want. We do the same when using a class. once we define a class, we can use it as many times as we want to create as many of the objects the class defines as we want. 

**Attributes** - Attributes are simply the description of an object. What makes up the object? We will look at a Class which defines cars in lecture. With a car, the attributes may be make, model, year and color. These are items we can use to describe every car we create.

**Methods** – Methods will be used to represent actions. In our example, the methods for a car may be to drive and stop. These are actions which all cars can perform.

To quickly review, we will create a Class to create specific object(s) (think of it as a blueprint, or cookie cutter). Within the class we will define attributes for object which describe the object and finally we will create some methods which represents what the object can do.  



```python
from math import sqrt
class Square:
    def __init__(self,name,size):
        self.name = name
        self.size = size
        
    def perimeter(self):
        return self.size * 4
    
    def area(self):
       return self.size **2

    def diag(self):
        return sqrt(2) * self.size
```

Looking at the code block above, it is a definition of a class which creates a square and with two attributes, name and size. The method(function) __init__() is known as a constructor. The constructor is repsonsible for creating new objects and assigning the attributes (name , size) to the object. 
The remining functions are the methods this class will contain, namely perimeter(), area(), diag() 
notice this class requires the use of a function not included in Python by default so we must use the import statment and import the sqrt funtcion. 

Our methods in this class allow us to calculate values for each of the method and return the values to the main program. 
To use this class file, we have two choices. 

1. Save this file as its own file and import the class into our main file. For example, if we save this file as square.py, in our main Python file, we would use the import statment as follows "from square.py import Square"

2. Our saecond option is to include the class declaration into our main program. While this works perfectly, It becomes very inconvenient to resue the class code. If we are looking to create reusable code, the best way to do so is to use a separate file to store the class.

### API - Applciation Programming Interface

An API is a layer of code which acts like a bridge to allow systems to talk to each other. In our case we want to talk to a source of data by using python, we need to use something in common or a pathway to get from our Python code to the data source. What is in common is the data source is on the Internet and we have access to the Internet via Python. An API can take advantage of the Representational State Transfer (REST) architectural style to access data and functions of a particular data source or website. The data and function are represented as resources. REST accesses resources through Uniform Resource Identifiers (URIs). A common language spoken on the Internet is HTTP and REST can use it as well.  An API is considered RESTful, if it adheres to the guiding principles of REST. We refer to APIs as RESTFUL if we can use it to access a data source via HTTP.

Particularly we need to be able to communicate by using HTTTP’s specific verbs

HEAD

GET

POST

PUT

PATCH

DELETE

 

So now we have a way to access a data source on the Internet via HTTP Verbs (REST), we just need to enable Python to be able to communicate via HTTP.  We only need to import one library in python, called “requests”

Importing requests into our code will provide the ability to communicate with our RESTful API using HTTP Verbs

To access this data in a Pythonic way, we will use a package named requests. Requests allows us to send HTTP commands from Pyton as if the commands were coming from a browser. We setup requests in a particular way. We also must send a header statement along with the URL for the API so that the listening API thinks it is communicating with a browser. If repeated requests come in to an API without headers, the API may think it is under attack and take measures to shut down or block the source IP address
The code block below sets up the request package to send a request to the open-notify API. Printing the returning results of a successful request will produce the Response 200 message. A 200 message indicates the request was successful and completed correctly

```python
import requests
headers = {"Content-Type":"application/json", "User-agent":"Mozilla"}
results = requests.get('http://api.open-notify.org/astros.json', headers=headers)
print(results)
```

To access the data returned, we send results to the JSON() method and convert the returning data to JSON, This will allow us to parse the JSON and pick out the items we want to display. 

```python
myJSON  = results.json()
```

To access the names in the data, the structure of the returning JSON must be viewed to understand how to get the correct piece if data to display.
Looking into the JSON formatter output, we see the retuning JSON loosk just like a disctionary in Python. The first key pair is "message":"success" so printing the message key should result in displaying the value ("success"). Notice the format of the print statement to print the value associated with the message key. 

```python
print (myJSON['message'])
```

There is a reserved attribute from the requests module. It is called .status_code. This simply returns the last status code from the request. The status code is useful to test if a page was successful

```python
print(results.status_code)   

if results.status_code != 200:
    print("Sorry there was something wrong with your request")
    
else:
    print("Hooray!, your request was a success")
    
```

To access the first person's name in Python we use the following syntax 

JSON\[key\]\[index\]\[key\]

Working from the outer most elements, JSON is the returning JSON, the first \[key\] is the first key we encounter , in this cale 'people'. The index is the position in the value of people (stored as a list). In our case, the first person is at index 0 and finally the next key represents the value of name inside the list inside the people key.
So in Python it looks like this


\['people'\]\[0\]\['name'\]

```python
print(myJSON['people'][0]['name'])
```
