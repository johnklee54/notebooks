## Chapter 6 Loops - Iteration
Loops are used to perform a task repeated until a condition has been satisified. We call this iteration. There are two basic loops in Python. 

* **While** loop - this loop performs actions within the loop body until a condition is satisfied. Picture this as combining the IF statement with a loop. it is conditional in the sense that the loop continues until a test is satisfied. This is useful if we want to keep our programs running until a user inputs a choice to quit
The general format of the while loop:
while *some_test*:
    *take some action*
    
* **For loop** - We use the For loop to iterate through things like lists, dictionaries and even strings.  Notice in the code, the letters are printing out one on each line ? that is because the for loop is iterating through the string and storing each letter in the variable letter. 

The **range()** function is one of the most useful functions to iterate. It generates numbers within as specified range. Just like slicing, range takes 3 parameters
* start - this is where to start a range, defaults to 0
* stop - this is where to stop the range. as with slicing , the stop parameter is not inclusive
* step - how many steps to skip

If range() is specified with just one value i.e range(5) that is the stop value

```python
print("A basic While loop")
count=1
while count <=15:
    print(f"This is the count {count}")
    count +=1
    
print("\n")

print("A basic For loop to iterate through a string")

myString="INFO-233"
for letter in myString:  # letter is holding each character of myString
    print(letter)

print("\n")

print("Using range to count to 4 from 0")

for x in range(5):
    print(x)
    

print("\nGiving the user a choice to quit")    
while True:  
    myStr = input("Please enter a number or q to quit: ")
    if myStr == 'q':
        print("Thanks for using this program")
        break
    try:
        myNum= int(myStr) **2
        print(f"The number squared is {myNum}")
    except ValueError:
        print("You entered a value which is not a number.")
    
```

* intersection  use "&" or .intersection() - members common to both sets

* subset - .issubset() - whether one set is a subset of another





```python
mlbTeams ={
    'Baltimore Orioles':'Baltimore',
    'Boston Red Sox':'Boston',
    'New York Yankees':'New York City',
    'Tampa Bay Rays':'St. Petersburg',
    'Toronto Blue Jays':'Toronto',
    'Chicago White Sox':'Chicago',
   
    }

  

print(f"\nThe city for the Boston Red Sox is {mlbTeams['Boston Red Sox']}")

print("\n")
print(f"The keys for the dictionary {mlbTeams.keys()}")
print("\n")
print(f"The values for the dictionary {mlbTeams.values()}")
print("\n")
print(f"The key:value pairs for the dictionary {mlbTeams.items()}")
print("\n")
print(f"The length of the dictionary {len(mlbTeams)} pairs")
print("\n")
print(f"Remove the last items of the dictionary \"Chicago White Sox\" {mlbTeams.pop('Chicago White Sox')}")

print(f"The dictionary no longer has the key or value for \"Chicago WHitre Sox\" {mlbTeams.keys()} {mlbTeams.values()}")

print("\n")
print("Iterate through the keys")
for team in mlbTeams:
    print(team)
    
print("\n")
print("Iterate through the values")
for cities in mlbTeams.values():
    print(cities)
    
print("\nIterate through the keys and values at the same time")
for teams in mlbTeams:
    print (teams,':',mlbTeams[teams])
```

```python

print("Create three sets, one master, one even and one odd numbers from 0 to 50 using a for loop")

myMasterSet = set()
myOddSet = set()
myEvenSet = set()



for number in range(51):
    myMasterSet.add(number)
for number in range(0,51,2):
    myEvenSet.add(number)
for number in range(1,51,2):
    myOddSet.add(number)


print(f"Master Set has {len(myMasterSet)} members")
print(myMasterSet)
print("\n")
print(f"Even Set has {len(myEvenSet)} members")
print(myEvenSet)
print("\n")
print(f"Odd Set has {len(myOddSet)} members")
print(myOddSet)
```

```python
## intersection  use "&" or .intersection() - members common to both sets
## difference -,difference() members of the first set but not members of  the second set
## subset - .issubset() - whether one set is a subset of another

first_set = {1, 2, 3, 4, 5, 6}
second_set = {4, 5, 6, 7, 8, 9}
whole_set =set() # empty set for use later
print(f"first_set {first_set}")
print(f"second_set {second_set}")
print("\n")
print(f"The first set intersects with the second set {first_set.intersection(second_set)}")

print(f"The first set difference with the second set {first_set.difference(second_set)}")

print(f"The second set difference with the first set {second_set.difference(first_set)}")

print("\n")


if first_set.issubset(second_set):
    print("The first set is a subset of the second set")
else: 
    print("The first set is not a subset of the second set")
    

for number in first_set:
    whole_set.add(number)

for number in second_set:
    whole_set.add(number)
    
print(f"Whole_set {whole_set}")

if first_set.issubset(whole_set):
    print(f"The first set is a subset of the whole set {first_set}")
else: 
    print("The first set is not a subset of the second set")
```
